### One hot encoding with multi category

```python
# Let try to find frequent category in X1
data1.X1.value_counts().sort_values(ascending=False)
```

`As display above , we can select only more frequent category `

```python
# Let try with top 10 frequent category
top_10 = [x for x in data1.X1.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
def encoding(data,colum,top_list):
    for label in top_list:
        data[label] = np.where(data[colum]==label,1,0)
```

```python
encoding(data1,'X1',top_10)
```

```python
data1[['X1']+top_10].head(40)
```

```python
# Let Apply for all colums
```

```python
colums = ['X1','X2','X3','X4','X5','X6']
data2 = pd.read_csv("train.csv",usecols=colums)
data2
```

```python
def one_hot_encoding(df,colums,top_list):
    for label in top_list:
        df[colums+'_'+label] = np.where(df[colums] == label,1,0)
```

```python
def get_top_category(df,column,top_number):
    return [x for x in df[column].value_counts().sort_values(ascending=False).head(top_number).index]
```

```python
for colum in colums:
    top_10 = get_top_category(data2,colum,10)
    one_hot_encoding(data2,colum,top_10)
```

```python
data2
```

`Using frequency ,now we got only 57-6 new columns`

 Using frequency count for category, we can reduce feature count but we loss to train model for less frequent category.

### Mean encoding

```python
colums = ['X1','X2','X3','X4','X5','X6','y']
data3 = pd.read_csv("train.csv",usecols=colums)
data3
```

```python
data3.isnull().sum()
```

```python
def mean_encoding(df,column):
    column_index = df[column].value_counts().index
    # create new column
    df[column+'_new'] = df[column].copy()
    for cat in column_index:
        category = df[df[column+'_new']== cat]
        mean = round(category.y.mean(),2)
        df[column+'_new'] = np.where(df[column+'_new'] == cat,mean,df[column+'_new'])
    
```

```python
new_colums = colums[:-1]
new_colums
```

```python
for col in new_colums:
    mean_encoding(data3,col)
```

### Label Encoding

```python
import datetime
```

```python
# create variable ith date and find the weekday
# Create 20 days dataframe
df_base = datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(20)]
df = pd.DataFrame(df_date_list)
df.columns = ["day"]
df
```

```python
df["day_of_week"] = df.day.dt.day_name()
df
```

```python
# create dis to assign order for day
weekday_map = {
    'Monday':1,
    "Tuesday":2,
    "Wednesday":3,
    "Thursday":4,
    "Friday":5,
    "Saturday":6,
    "Sunday":7
}
```

```python
# Assign order for all day
df["day_order"] = df.day_of_week.map(weekday_map)
df.head(20)
```

## Target guided Ordinal Encoding

```python
df = pd.read_csv("titanic.csv",usecols=['Cabin','Survived'])
df
```

```python
df.Cabin.fillna('Missing',inplace=True)
```

```python
df.head(20)
```

```python
df.Cabin = df.Cabin.astype(str).str[0]
```

```python
df
```

```python
df.Cabin.unique()
```

```python
original_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
original_labels
```

```python
original_labels2 = {k:i for i,k in enumerate(original_labels,0)}
original_labels2
```

```python
df['Cabin_ordinal_labels'] = df.Cabin.map(original_labels2)
df
```
